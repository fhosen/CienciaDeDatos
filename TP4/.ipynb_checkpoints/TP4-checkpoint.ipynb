```python
import scipy.signal as sig
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
tuviejanadamas = pd.read_csv("data.csv", names = ["Años", "LetrasYNros", "Areas", "Autores", "Nombres"], error_bad_lines=False)
tuviejanadamas
```

```python
def authors_from(index):
    authors_str = tuviejanadamas["Autores"][index]
    try: return authors_str.split("&")
    except AttributeError:
        print (tuviejanadamas["Autores"][index])
        return []
    

def get_all_authors():
    all_authors = []
    for index in range(tuviejanadamas.shape[0]):
        all_authors += authors_from(index)
    return set(all_authors)

def edges_from(elems):
    for e1 in elems:
        for e2 in elems:
            yield e1, e2

def graph_of_collaborations():
    graph = nx.Graph()
    all_authors = get_all_authors()
    graph.add_nodes_from(all_authors)
    
    for index in range(tuviejanadamas.shape[0]):
        authors = authors_from(index)
        if len(authors) > 1:
            permutations = edges_from(authors)
            graph.add_edges_from(permutations)
    
    return graph, all_authors 

graph, all_authors = graph_of_collaborations()
subgraph = graph.subgraph(list(all_authors)[1:500])

plt.figure
nx.draw_networkx_edges(subgraph, node_size=1)
plt.show()

```
